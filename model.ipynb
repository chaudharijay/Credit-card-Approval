```python
import pandas as pd
```

```python
df = pd.read_csv('Credit_card.csv')
df.shape
```

```python
label = pd.read_csv("Credit_card_label.csv")
label
```

```python
df  = df.merge(label, on ='Ind_ID', how="left") # doing left join
df
```

```python
average_income = df['Annual_income'].mean()
average_income
```

```python
gender_income = df.groupby("GENDER")['Annual_income'].mean()
gender_income
```

```python
gender_income.sort_values(ascending=False)
```

```python
gender_income.sort_values(ascending=False)[0]
gender_income.sort_values(ascending=False)[1]
```

```python
sns.barplot(x=gender_income.index, y=gender_income.values)
plt.title("Average Income by Gender")
plt.xlabel("Genders")
plt.ylabel("Average Income")
plt.legend()
```

```python
label_counts = df['label'].value_counts()
label_counts
```

```python
plt.pie(label_counts, labels=['Credit Approved', "Credit Rejected"], autopct="%1.1f%%")
plt.title("Credit Approval Status")
plt.show()
```
